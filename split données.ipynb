```python
import pandas as pd
import os
from pathlib import Path

def split_csv(file_path, output_dir, lines_per_file=50):

    df = pd.read_csv(file_path)
    

    total_lines = len(df)
    

    num_files = total_lines // lines_per_file + int(total_lines % lines_per_file > 0)
    

    Path(output_dir).mkdir(parents=True, exist_ok=True)
    
    for i in range(num_files):
        start_index = i * lines_per_file
        end_index = start_index + lines_per_file
        df_chunk = df.iloc[start_index:end_index]
        

        chunk_file_name = os.path.join(output_dir, f"_part_{i+1}.csv")
        df_chunk.to_csv(chunk_file_name, index=False)
        print(f"File {chunk_file_name} created with lines from {start_index} to {end_index - 1}")


file_path = 'Disruptions_Dataset.csv'
output_dir = 'input-split'
split_csv(file_path, output_dir)
```
